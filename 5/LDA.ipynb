```python
import re
import string
from gensim import corpora, models
import jieba
import pandas as pd

# 讀取文本文件
with open('tp.txt', 'r', encoding='utf-8') as file:
    text_content = file.read()

# 進行清理與預處理
text_content = text_content.replace('的', '').replace('在', '').replace('是', '').replace('與', '').replace('對', '').replace('於', '').replace('為', '').replace('而', '')
pattern = r'[ \u4e00-\u9fa5 ]+'
text_content = re.findall(pattern, text_content)
text_content = ''.join(text_content)

# 使用 jieba 分詞
seg_list = jieba.lcut(text_content)

# 創建字典與語料庫
dictionary = corpora.Dictionary([seg_list])
corpus = [dictionary.doc2bow(seg_list)]

# 訓練 LDA 模型
lda_model = models.LdaModel(corpus, num_topics=16, id2word=dictionary, passes=10)

# 輸出主題模型分析結果
topics = lda_model.print_topics(num_words=8)
for topic in topics:
    print(topic)
```
