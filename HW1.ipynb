```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns

# 讀取檔案
df1 = pd.read_csv('nba0001.csv')

# 分布圖
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df1, x='team', y='rate')

# 添加標題
plt.title('nba00-01')

plt.show()
```

```python
# 讀取檔案
df2 = pd.read_csv('nba2223.csv')

# 分布圖
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df2, x='team', y='rate')

# 添加標題
plt.title('nba22-23')

plt.show()
```

```python
# 統計rate大於0.5的比例
rate1_05 = (df1['rate'] > 0.5).mean()

# 圓餅圖
labels = ['Rate > 0.5', 'Rate <= 0.5']
sizes = [rate1_05, 1 - rate_05]
colors = ['#66b3ff','#ff9999']
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.0f%%', startangle=90)
ax1.axis('equal')

# 添加標題
plt.title('Rate Distribution 00-01')

plt.show()
```

```python
# 統計rate大於0.5的比例
rate2_05 = (df2['rate'] > 0.5).mean()

# 圓餅圖
labels = ['Rate > 0.5', 'Rate <= 0.5']
sizes = [rate2_05, 1 - rate_05]
colors = ['#66b3ff','#ff9999']
 
fig2, ax2 = plt.subplots()
ax2.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=270)
ax2.axis('equal')

# 添加標題
plt.title('Rate Distribution 22-23')

plt.show()
```

```python
# 找到檔案中的隊伍名稱
teams_df1 = set(df1['team'].unique())
teams_df2 = set(df2['team'].unique())

# 找到相同的隊伍
common_teams = teams_df1.intersection(teams_df2)

# 找到不同的隊伍
different_teams = teams_df1.symmetric_difference(teams_df2)

# 輸出结果
print("相同的隊伍：", common_teams)
print("不同的隊伍：", different_teams)
print("相同隊伍的数量：", len(common_teams))
print("不同隊伍的数量：", len(different_teams))
```

```python
# 計算相同和不同隊伍的数量
common_teams_count = len(common_teams)
different_teams_count = len(different_teams)

# 條狀圖
plt.figure(figsize=(8, 5))
plt.bar(['common', 'different'], [common_teams_count, different_teams_count], color=['purple', 'yellow'])

# 添加標題
plt.title('amount')

plt.show()
```
